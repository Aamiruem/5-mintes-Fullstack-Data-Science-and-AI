You're getting this error because you're trying to access an index that does not exist in the 3rd axis (axis=2) of your NumPy array.

â— Error meaning
IndexError: index 1 is out of bounds for axis 2 with size 1


This means:

Axis 2 (the third dimension) has size 1

You are trying to access index 1

Valid index is only 0

So a4[0,0,1,1] is invalid because the 3rd dimension index = 1 is out of range.

âœ… How to fix it?

You must check the shape of your array a4 first:

print(a4.shape)


Suppose you get something like:

(2, 2, 1, 2)


This means:

axis 0 â†’ size 2

axis 1 â†’ size 2

axis 2 â†’ size 1 â† â— problem

axis 3 â†’ size 2

So valid indices:

Axis	Size	Valid indices
0	2	0â€“1
1	2	0â€“1
2	1	0 only
3	2	0â€“1
âœ” Correct valid indexing example
a4[0, 0, 0, 1]   # axis 2 index must be 0

ðŸŽ¯ If you want axis 2 to have size > 1

Your array must contain more elements, for example:

a4 = np.random.randint(10, size=(2,2,2,2))
print(a4.shape)
# (2, 2, 2, 2)


Now this will work:

a4[0,0,1,1]

ðŸ“Œ Summary

You are accessing index 1 on axis 2 but that axis has size 1, so index 1 does not exist.
Only index 0 is allowed for that axis.
