```python
import numpy as np

a1 = np.array([1, 2, 3])

# a1.ndim → tells the number of dimensions (axes) of the array.
# For example: 
# 1D array → ndim = 1  (like [1,2,3])
# 2D array → ndim = 2  (like [[1,2,3],[4,5,6]])
# 3D array → ndim = 3  (like [[[1,2],[3,4]]])
print(a1.ndim)
```

```python
a2 = np.array([
    [1, 2, 3],
    [10, 20, 30],
    [100, 200, 300],
    [1000, 2000, 3000]
])

# This creates a 2D NumPy array (matrix) with 4 rows and 3 columns:
# [
#   [1,     2,     3    ],
#   [10,    20,    30   ],
#   [100,   200,   300  ],
#   [1000,  2000,  3000 ]
# ]
# → Shape = (4, 3)
# → ndim = 2 (because it is a 2D array)
# → Each inner list represents one row
```
